```python
import pandas as pd
selloutData=pd.read_csv("./dataset/kopo_channel_seasonality_new.csv")
selloutData.head()
```

```python
## PIVOT(행을 열로)
## pivot : 데이터프레임명.pivot_table( index = [‘고정key컬럼1’,’고정key컬럼2’], \
## columns=‘pivot대상컬럼명’, values=‘피벗산대상컬럼’, aggfunc=“mean”)
```

```python
pivotResult=selloutData.\
pivot_table(index=['REGIONID','PRODUCT'],columns='YEARWEEK',values='QTY',aggfunc="mean")
pivotResult2=pivotResult.reset_index()
```

```python
pivotResult2.head()
```

```python
## UNPIVOT(열을 행으로)
##unpivot : pd.melt(데이터프레임명, id_vars = [‘컬럼#1’,’컬럼#2’], var_name='YEARWEEK', value_name='QTY') 
```

```python
unpivotResult=pd.melt(pivotResult2,id_vars=['REGIONID','PRODUCT'],var_name='YEARWEEK', value_name='QTY')
unpivotResult.head()
```
