```python
groupKey=["lvy_rper_tin","le_tin"]
meanValue=texdata1.groupby(groupKey,as_index=True).sum()
meanValue
```

```python
meanValue=meanValue.reset_index()
meanValue2=meanValue[meanValue.cnt>1]
meanValue3=meanValue2["le_tin"]
meanValue3=meanValue3.reset_index(drop=True)
meanValue3
```

```python
meanValue4=meanValue2.le_tin.tolist()
meanValue4
```
