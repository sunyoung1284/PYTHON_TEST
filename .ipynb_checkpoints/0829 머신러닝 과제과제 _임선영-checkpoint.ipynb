```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.model_selection import train_test_split
```

```python
selloutData=pd.read_csv("https://raw.githubusercontent.com/hyokwan/python-lecture/master/dataset/WA_Fn-UseC_-Sales-Win-Loss.csv")
```

```python
selloutData.head()
```

```python
selloutData.columns
```

```python
selloutData.columns = ['Opportunity_Number', 'Supplies_Subgroup', 'Supplies_Group', 'Region',
       'Route_To_Market', 'Elapsed_Days_In_Sales_Stage', 'Opportunity_Result',
       'Sales_Stage_Change_Count', 'Total_Days_Identified_Through_Closing',
       'Total_Days_Identified_Through_Qualified', 'Opportunity_Amount_USD',
       'Client_Size_By_Revenue', 'Client_Size_By_Employee_Count',
       'Revenue_From_Client_Past_Two_Years', 'Competitor_Type',
       'Ratio_Days_Identified_To_Total_Days',
       'Ratio_Days_Validated_To_Total_Days',
       'Ratio_Days_Qualified_To_Total_Days', 'Deal_Size_Category']
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
selloutData['LE_Supplies_Subgroup'] = le.fit_transform(selloutData.Supplies_Subgroup)
selloutData['LE_Supplies_Group'] = le.fit_transform(selloutData.Supplies_Group)
selloutData['LE_Region'] = le.fit_transform(selloutData.Region)
selloutData['LE_Route_To_Market'] = le.fit_transform(selloutData.Route_To_Market)
selloutData['LE_Opportunity_Result'] = le.fit_transform(selloutData.Opportunity_Result)
selloutData
```

```python
corrDf=selloutData.corr()
corrDf
```

```python
featuresStd = 0.5

features = list(corrDf[abs(corrDf.LE_Opportunity_Result) > featuresStd].index)
features

```

```python
features=  ["LE_Supplies_Subgroup", "LE_Supplies_Group", "LE_Region", "LE_Route_To_Market", "Opportunity_Amount_USD"]
label= ['Opportunity_Result']
```

```python
trainingData_features,testData_features,trainingData_label,testData_label = train_test_split(selloutData[features],selloutData[label],test_size = 0.3,random_state = 3)
```

```python
model_method = svm.SVC()
```

```python
model = model_method.fit(trainingData_features,trainingData_label)
```

```python
predict=model.predict(testData_features)
predict
```

```python
testData_features = testData_features.reset_index(drop=True)
testData_features
finalResult1 = pd.concat([testData_features.reset_index(drop=True),testData_label.reset_index(drop=True)], axis=1)
```

```python
len(finalResult1)
```

```python
len(predict)
```

```python
finalResult1["PREDICT"] = predict
```

```python
finalResult1
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
```

```python
# 결과 테스트하기 
ac_score = accuracy_score(finalResult1.Opportunity_Result, finalResult1.PREDICT)
cl_report = classification_report(finalResult1.Opportunity_Result, finalResult1.PREDICT)

## 결과 리포트하기
print("Accuracy =", ac_score)
print("result =\n", cl_report)
```
