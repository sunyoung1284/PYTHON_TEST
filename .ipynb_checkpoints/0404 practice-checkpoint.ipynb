```python
import pandas as pd
selloutData=pd.read_csv('./dataset/kopo_channel_seasonality_new.csv')
selloutData.head()
```

```python
selloutData.dtypes
```

```python
selloutData["QTY"]=selloutData["QTY"].astype(float)
selloutData["YEARWEEK"]=selloutData["YEARWEEK"].astype(str)
selloutData["PRODUCT"]=selloutData["PRODUCT"].astype(str)
selloutData["REGIONID"]=selloutData["REGIONID"].astype(str)
```

```python
selloutData.dtypes
```

```python
import numpy as np
```

```python
selloutData["QTY_NEW"]=np.where(selloutData["QTY"]<0,0,selloutData["QTY"])
selloutData.head()
```

```python
selloutData["YEAR"]=selloutData.YEARWEEK.str[0:4]
selloutData["WEEK"]=selloutData.YEARWEEK.str[4:]
```

```python
selloutData["WEEK"]=selloutData["WEEK"].astype(int)
```

```python
refinedSelloutData=selloutData[selloutData.WEEK<=52]
refinedSelloutData.head()
```

```python
sortedData=selloutData.sort_values(["REGIONID","PRODUCT","YEARWEEK"])
sortedData.head()
```

```python
groupData=sortedData.groupby(['REGIONID','PRODUCT','YEAR']).agg(['mean','std'])["QTY_NEW"]
groupData=groupData.reset_index().head()
```

```python
selloutData.head()
```

```python
groupData.columns=["REGIONID","PRODUCT","YEAR","QTY_MEAN","QTY_STD"]   
groupData.head()
```

```python
mergedData=pd.merge(selloutData,groupData,left_on=["REGIONID","PRODUCT","YEAR"],right_on=["REGIONID","PRODUCT","YEAR"],how="left")\
[["REGIONID","PRODUCT","YEARWEEK","YEAR","WEEK","QTY_NEW","QTY_MEAN","QTY_STD"]]
```

```python
mergedData.head()
```
