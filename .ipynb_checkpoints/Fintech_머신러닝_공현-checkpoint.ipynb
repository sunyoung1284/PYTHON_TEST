```python
import numpy as np
import pandas as pd 
from sklearn import datasets, tree

kdtan = pd.read_csv('Python_Fintech/Python_ST_EX/dataset/kopo_decision_tree_all_new.csv')
kdtan['HO_YN'] = np.where(kdtan.HOLIDAY == 'Y', 1, 0)
kdtan['PRO_YN'] = np.where(kdtan.PROMOTION == 'Y', 1, 0)

def ML(G):
    
    if len(G) >= 63 :
        
        corrDf = G.corr()
        corrDf = corrDf[(abs(corrDf.QTY) >= 0.01) & (corrDf.QTY != 1)]
        
        label = ['QTY']
        features = list(corrDf.index)
        
        G = G.sort_values(['YEARWEEK']).reset_index(drop = True)
        
        indexStd = round(len(G) * 0.8)
        
        trainingData = G.loc[:indexStd]
        testData = G.loc[indexStd:]
        
        trainingData_features = trainingData[features]
        trainingData_label = trainingData[label]
        
        testData_features = testData[features]
        testData_label = testData[label]
        
        model_method = tree.DecisionTreeRegressor(random_state = 1)
        model = model_method.fit(trainingData_features, trainingData_label)
        
        predict = model.predict(testData_features)
        predictData = pd.DataFrame(predict)
        
    else:
        
        testData = G
        predictData = pd.DataFrame([0 for i in range(len(G))])
        
    predictData.columns = ['PREDICT']
    G = pd.concat([testData.reset_index(drop = True), predictData.reset_index(drop = True)], axis = 1)
    
    return G

result = kdtan.groupby(['REGIONID', 'PRODUCTGROUP', 'PRODUCT', 'ITEM']).apply(ML)
result
```
