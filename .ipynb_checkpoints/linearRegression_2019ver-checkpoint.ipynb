### 1. 데이터 불러오기

```python
#CSV 파일 불러오기
selloutData = pd.read_csv('KOPO_DECISION_TREE_ALL_NEW.csv')
# 컬럼해더 재정의
selloutData.columns = selloutData.columns.str.upper()
# 데이터 VIEW
selloutData.head()
```

### 2. 데이터 필터 및 정렬

```python
# 컬럼삭제
#del(selloutData["HCLUS"])
```

```python
# 오름차순정렬
selloutData.sort_values(["REGIONID","PRODUCTGROUP","PRODUCT","ITEM","YEARWEEK"], 
                        ascending=True,inplace=True)
#인덱스 초기화
selloutData.reset_index(drop=True,inplace=True)
```

```python
#뉴메릭 타입으로 변환
a = {"HOLIDAY" : {"Y" : 1 , "N" : 0}}
selloutData.replace(a,inplace =True)
```

```python
groupKey = ["REGIONID","PRODUCTGROUP","PRODUCT","ITEM"]
```

```python
selloutData.groupby(by = groupKey).describe()
```

```python
#그룹바이 필터  -- 146주가 다포함된 데이터만 추출
#그룹바이 끝날때마다 그룹이 되어 인덱스가 변하기때문에 reset_index를 항상해줘야함
selloutData=selloutData.groupby(by = groupKey).filter(lambda x: x["YEARWEEK"].count()==146).reset_index(drop=True)
```

### 3. 머신러닝 linear Regression 적용

```python
def sub_function(data):
    #그룹바이 다음에 바로 apply(sub_function)이 들어갈거기때문에 맨첫줄에 reset_index 
    data.reset_index(inplace=True,drop=True)
    #####################train,test 데이터 분리###################################
    trainData = data[data["YEARWEEK"].astype(int)<201700].reset_index(drop = True)
    testData = data[data["YEARWEEK"].astype(int)>=201700].reset_index(drop = True)
    testData['YEARWEEK'] = testData['YEARWEEK']
    train_feature = trainData[ ["WEEK","HOLIDAY",'PRO_PERCENT']]
    #train_feature = trainData[ ["YEARWEEK","HOLIDAY",'PRO_PERCENT']]
    #train_feature = trainData[ ["HOLIDAY",'PRO_PERCENT']]
    train_label = trainData[['QTY']]
    test_feature = testData[ ["WEEK","HOLIDAY",'PRO_PERCENT']]
    #test_feature = testData[ ["YEARWEEK","HOLIDAY",'PRO_PERCENT']]
    #test_feature = testData[ ["HOLIDAY",'PRO_PERCENT']]
    #################linearRegression 모델 적용 ##########################
    lm = linear_model.LinearRegression()
    model = lm.fit(train_feature, train_label)
    predict = lm.predict(test_feature)
    predictDF = pd.DataFrame(data = predict , columns=["PREDICT"])
    testData["PREDICT"]=predictDF
    return testData
```

```python
#그룹바이 traindata ,testdata 는 yearweek 분류 < - 피처값으로는 week
#yearweek으로 추세선 설정시 ex2100년기준으로보면 하락추세면 계속 하락하기때문에 음수값으로만 나온다. 
#week으로 설정시 평균값으로 추세를 내서 구하는것같음.
```

```python
newselloutData = selloutData.groupby(["REGIONID","PRODUCTGROUP","PRODUCT","ITEM"]).apply(sub_function).reset_index(drop=True)
```

### 4. groupby 디버깅 시각화

```python
newselloutData2 = newselloutData.groupby(["REGIONID","PRODUCTGROUP","PRODUCT","ITEM"])
```

```python
#실제값과 예측값 시각화 비교
plt.plot(eachGroup['YEARWEEK'],eachGroup['QTY'])
plt.plot(eachGroup['YEARWEEK'],eachGroup['PREDICT'])
```
