```python
import pandas as pd
import numpy as np
import math
#모델 라이브러리 선언
from sklearn import datasets, tree
#모델 정확도 라이브러리 선언
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
```

```python
featuresData=pd.read_csv("./dataset/feature_regression_example.csv")
```

```python
featuresData.head()
```

```python
featuresData.dtypes
featuresData["PRO_YN"] = np.where(featuresData.PROMOTION=="Y",1,0)
featuresData["H_YN"] = np.where(featuresData.HOLIDAY=="Y",1,0)
featuresData.head()
```

```python
#get_dummies
pd.concat([featuresData,pd.get_dummies(prefix="PROMO",data = featuresData.PROMOTION)], axis =1)
```

```python
#sklearn
from sklearn.preprocessing import LabelEncoder
le_pro = LabelEncoder()
le_holy =LabelEncoder()
featuresData['HOLY_ENCO']= le_pro.fit_transform(featuresData.PROMOTION)
```

```python
le_pro.inverse_transform(featuresData['HOLY_ENCO'])  ##되돌리기
```

```python
featuresData.head()
```

```python
#dictionary mapping
```

```python
binarymap = {"Y":1,"N":0}
featuresData["PRO_DICT"] = featuresData.PROMOTION.map(binarymap)
featuresData.head(2)
```

```python
###특정선정 및 데이터 분리
#상관관계 확인
corrDf=featuresData.corr()
corrDf
```

```python
featuresStd = 0.5

features = list(corrDf[abs(corrDf.QTY) > featuresStd].index)
features

label = ['QTY']
label
```

```python
features
```

```python
#데이터 분리
```

```python
indexStd = round(len(featuresData)*0.8)
```

```python
featuresData.iloc[0:indexStd,:]
```

```python
yearweekStd = featuresData.YEARWEEK[indexStd]
yearweekStd
```

```python
trainingData_features = featuresData.iloc[0:indexStd][features]
```

```python
trainingData_features = featuresData[featuresData.YEARWEEK <= yearweekStd][features]
trainingData_label = featuresData[featuresData.YEARWEEK <= yearweekStd][label]
testData_features = featuresData[featuresData.YEARWEEK > yearweekStd][features]
testData_label = featuresData[featuresData.YEARWEEK > yearweekStd][label]
```

```python
###   DecisionTree   ###
# Define Model
model_method = tree.DecisionTreeRegressor(random_state=1)

# Learning (Feature & Label)
model = model_method.fit(trainingData_features, trainingData_label)

# 예측
predict = model.predict(testData_features)
predict
```

```python
### 데이터프레임으로 변환
predictData = pd.DataFrame(predict)

### 컬럼명 주입
predictData.columns = ["PREDICT"]
predictData.head()
```

```python
###  테스트 데이터와 예측값을 붙이세요  ###   
testData = featuresData[featuresData.YEARWEEK > yearweekStd][label]
testDataFinal = testData.reset_index(drop=True)
finalResult = pd.concat([testDataFinal, predictData], axis = 1)
finalResult.head(2)
```

```python
###  RandomForest   ### 
```

```python
forest = RandomForestClassifier(n_estimators=100)
forest.fit(trainingData_features, trainingData_label)
RandomForestPredict = forest.predict(testData_features)
print(RandomForestPredict)
```

```python
predictData2 = pd.DataFrame(RandomForestPredict)

predictData2.columns = ["PREDICT"]
predictData2.head()
```

```python
finalResult2 = pd.concat([testDataFinal, predictData2], axis = 1)
finalResult2.head(5)
```

```python
#######   정확도 구하기    #######
### MAE
mean_absolute_error(finalResult['QTY'], finalResult['PREDICT'])
```
